```python
import matplotlib.pyplot as plt
import numpy as np

# Models and metrics
models = ["BERT-base", "BERT-large", "BERT-large + Prompt"]
accuracy = [43.1, 44.7, 45.4]        # Accuracy in %
macro_f1 = [0.427, 0.443, 0.450]    # Macro F1

# Baseline metrics
baseline_acc = 33.7
baseline_f1 = 0.242

# Bar width and positions
x = np.arange(len(models))
width = 0.35

plt.figure(figsize=(8,5))

# Bars for Accuracy and F1
bars_acc = plt.bar(x - width/2, accuracy, width, label='Accuracy (%)', color='#1f77b4')
bars_f1 = plt.bar(x + width/2, [f*100 for f in macro_f1], width, label='Macro F1 (%)', color='#ff7f0e')  # scaled to % for display

# Baseline horizontal lines
plt.axhline(y=baseline_acc, color='red', linestyle='--', linewidth=2, label=f'Baseline Accuracy: {baseline_acc}%')
plt.axhline(y=baseline_f1*100, color='green', linestyle='--', linewidth=2, label=f'Baseline Macro F1: {baseline_f1:.3f}')

# Add data labels on top of bars
for bar in bars_acc:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval:.1f}%', ha='center', va='bottom', fontsize=10)
for bar in bars_f1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval:.1f}%', ha='center', va='bottom', fontsize=10)

# Labels, title, legend
plt.ylabel("Percentage (%)")
plt.title("BERT Model Accuracy and Macro F1 vs Baseline on ANLI")
plt.xticks(x, models)
plt.ylim(0, 55)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

plt.tight_layout()
plt.show()
```
